```python
def calculate_IQR(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    return (lower_limit, upper_limit)
```

```python
housing.select_dtypes([float]).apply(calculate_IQR, axis=0) 
```

```python
df_filtered = housing.select_dtypes([float]).apply(lambda x: x[(x >= calculate_IQR(x)[0]) & (x <= calculate_IQR(x)[1])])
housing_filtered = housing.loc[df_filtered.dropna().index]
```

```python
housing_filtered.shape
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
housing_filtered.hist(ax=ax, bins=50)
plt.show()
```

Neste caso, foi removido os outliers via IQR, afim dos mesmos não inviesar o modelo, por conta de valores fora do range dos percentis.

#### 1) A média de idade da casa interfere no preço médio da casa?

```python
housing['mean_age_cat'] = pd.cut(housing['housing_median_age'], bins=[1, 18, 29, 37, 52], include_lowest=True,
                                labels=["1-18", "18-29", "29-37", "37-52"])

housing['mean_age_cat2'] = pd.cut(housing['housing_median_age'], bins=[1, 18, 37, 52], include_lowest=True,
                                labels=["1-18", "18-37", "37-52"])
```

```python
housing.head()
```

```python
grouped_by_age = housing.groupby('mean_age_cat').mean().sort_values(by='median_house_value', ascending=False)
grouped_by_age_2 = housing.groupby('mean_age_cat2').mean().sort_values(by='median_house_value', ascending=False)
```

```python
fig = plt.figure(figsize = (20, 15))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)

ax1.plot(grouped_by_age.index, grouped_by_age.median_house_value)
ax2.plot(grouped_by_age_2.index, grouped_by_age_2.median_house_value)
```

Observa-se que o valor médio das casas por idade tende a aumentar de forma linea tem uma média maior de idade da residência;

dos 18 aos 37 anos os valores tendem a ter uma constância (gastos com filhos pode ser uma justificativa para não ter um aumento ou estudos/profissão mais estabilizada)

#### 2) Total de quartos e cômodos

Estamos criando novas colunas para quarto, cômodos e tamanho da casa. Convertendo square feet para m2.

para realizar a conversão é simplesmente pegar o tamanho em pés e dividir por 10.764

```python
housing['total_rooms_m2'] = housing.total_rooms/10.764
housing['total_bedrooms_m2'] = housing.total_bedrooms/10.764
housing['total_size_m2'] = housing.total_rooms_m2 + housing.total_bedrooms_m2
```

```python
housing[housing.total_size_m2 >= 518.00]['total_size_m2'].count()
```

10% da base tem casas com mais que 518m2, sendo um total estimado de 1960 registros

```python
def classify_house_m2(m2):
    if m2<=100:
        return '<100m'
    elif m2 > 100 and m2 <= 200:
        return '100m-200m'
    elif m2 > 200 and m2 <= 300:
        return '200m-300m'
    elif m2 > 300 and m2 <= 400:
        return '300m-400m'
    elif m2 > 400 and m2 <= 10000:
        return '>400m'
    else:
        return '-'
```

```python
housing['cat_m2'] = housing.total_size_m2.apply(classify_house_m2)
```

```python
housing
```

```python
housing[['cat_m2']].value_counts()
```

```python
housing.groupby('cat_m2').median()['median_house_value'].plot(kind='bar')
```

como possuem um número de casas menores em maior quantidade, gráficamente vemos os gráficos próximos.

Quanto maior o m2, maior o preço médio da casa.

## Regressão Linear

#### Feature Importance

```python
X = housing_filtered.drop(columns=['median_house_value']).copy()
y = housing_filtered.median_house_value.copy()
```

```python
for feature in X.columns:
    corr, p_value = pearsonr(housing_filtered[feature], y)
    print(f'{feature}: p-value={p_value}, correlation={corr}')
```

Population é a única feature que foi maior que 0.05, porém visto o problema de negócio para descobrir o valor das casas, acredito que seja interessante manter.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30)
```

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

## Métricas

```python
diff = y_test - y_pred
sq_diff = diff**2

mse = sq_diff.mean()
print("mse score:", mse)

rmse = np.sqrt(mse)
print("rmse score:", rmse)

r2 = r2_score(y_test, y_pred)
print("R2 score:", r2)

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```
